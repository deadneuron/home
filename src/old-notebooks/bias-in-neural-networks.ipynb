# What is the role of bias in neural networks?

Bias is an essential component of artificial neural networks (ANNs) that allows them to learn complex patterns. It plays a crucial role in shifting the activation function, thus controlling the neuron's output. The bias term can be thought of as the intercept in a linear regression model, which allows the model to fit the data better by translating the decision boundary.

$$
y = f(w_1 x_1 + ... + w_n x_n + b)
$$

where $y$ is the neuron's output, $f$ is an activation function, $w_i$ are the weights, $x_i$ are the inputs, and $b$ is the bias term.

In the context of artificial neurons, bias helps in achieving better flexibility and adaptability while learning. Without the bias term, the neuron's output would always pass through the origin, limiting the network's ability to model complex relationships.

```python
import numpy as np
import matplotlib.pyplot as plt

def activation_function(x):
    return 1 / (1 + np.exp(-x))

def neuron_output(inputs, weights, bias):
    weighted_sum = np.dot(inputs, weights) + bias
    return activation_function(weighted_sum)

inputs = np.array([0.5, 0.6])
weights = np.array([0.1, -0.2])
bias_values = np.linspace(-2, 2, 100)
outputs = [neuron_output(inputs, weights, b) for b in bias_values]

plt.plot(bias_values, outputs)
plt.xlabel("Bias")
plt.ylabel("Output")
plt.title("Neuron Output vs Bias")
plt.grid()
plt.show()
```
